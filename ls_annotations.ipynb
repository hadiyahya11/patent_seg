```python
import pandas as pd
import numpy as np
import ast
import cv2
```

# Notebook to convert label studio annotations to format we want

```python
data2 = pd.read_csv('annotations_nc_on_preprocessed_alt_20241025.csv')
data2
```

```python
# Load the data
data = pd.read_csv('250_annot.csv')
data
```

```python
#add data2 to data
data = data.append(data2)
data
```

```python
columns = ['image, label']
#keep only the columns we need
data = data[['image', 'label']]

data

```

```python
data['image'] = data['image'].apply(lambda x: x[x.find('GB'):])
```

```python
data
```

```python
#print duplicates in the data
image_counts = data['image'].value_counts()
image_counts[image_counts > 1]
```

```python
#drop duplicates
data = data.drop_duplicates(subset='image', keep='last')
```

```python
len(data['image'].unique())
```

```python
data['label'] = data['label'].apply(ast.literal_eval)
```

```python
data = data.explode('label')
data
```

```python
# convert from LS percent units to pixels 
def convert_from_ls(result):
    if 'original_width' not in result or 'original_height' not in result:
        return None


    w, h = result['original_width'], result['original_height']

    if all([key in result for key in ['x', 'y', 'width', 'height']]):
        return (int(w * result['x'] / 100.0), \
               int(h * result['y'] / 100.0), \
               int(w * result['width'] / 100.0), \
               int(h * result['height'] / 100.0))
```

```python
data['label'] = data['label'].apply(convert_from_ls)
data
```

```python
data_path = './dataset-dev/plates_png/'
image_file= 'GB.505944.A-008.png'
image_path = data_path + image_file
image = cv2.imread(image_path)

df = data[data['image'] == image_file]
for i,r in df.iterrows():
    x,y,w,h = r['label']
    cv2.rectangle(image, (x,y), (x+w, y+h), (0, 255, 0), 2)


cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
data.to_csv('data_dl.csv', index=False)
```
